```python
def remove_duplicates(input_list):
    return list(dict.fromkeys(input_list))

list_with_duplicates = [1, 2, 2, 3, 4, 4, 5, 1, 6, 3]
unique_list = remove_duplicates(list_with_duplicates)

print(f"List with duplicates: {list_with_duplicates}")
print(f"List without duplicates: {unique_list}")
```
