```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

# Example usage
sample_string = "Hello, welcome to Python programming!"
vowel_count = count_vowels(sample_string)

print(f"String: '{sample_string}'")
print(f"Number of vowels: {vowel_count}")
```
