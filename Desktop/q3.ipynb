```python
person_info = {
    "name": "Alice",
    "age": 28,
    "city": "New York",
    "profession": "Software Engineer"
}

print("--- Keys ---")
for key in person_info.keys():
    print(key)

print("\n--- Values ---")
for value in person_info.values():
    print(value)

print("\n--- Keys and Values ---")
for key, value in person_info.items():
    print(f"{key}: {value}")
```
